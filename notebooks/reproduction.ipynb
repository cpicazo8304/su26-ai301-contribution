### Creating a Daft DataFrame of Strings

YouDaft provides several ways to create a DataFrame. To create a DataFrame from a list of strings, you can use `daft.from_pylist` where each dictionary in the list represents a row and its keys are column names. Alternatively, you can first create a pandas DataFrame and then convert it to a Daft DataFrame using `daft.from_pandas`.

```python
# Example 1: Using daft.from_pylist
data = [
    {"text": "Hello Daft"},
    {"text": "This is a string DataFrame"},
    {"text": "Another example"}
]
df_daft_pylist = daft.from_pylist(data)

print("Daft DataFrame created with daft.from_pylist:")
df_daft_pylist.show()
```

## Testing String Functions that are Available

```python
import daft.functions as F

# Using string functions on the 'text' column of df_daft_pylist

# Convert to uppercase
df_upper = df_daft_pylist.with_column('text_upper', F.upper(df_daft_pylist['text']))
print("Uppercase strings:")
df_upper.show()
```

```python
# Get string length
df_length = df_daft_pylist.with_column('text_length', F.length(df_daft_pylist['text']))
print("\nString lengths:")
df_length.show()
```

```python
# Check for containment
df_contains = df_daft_pylist.with_column('contains_daft', F.contains(df_daft_pylist['text'], 'Daft'))
print("\nStrings containing 'Daft':")
df_contains.show()
```

```python
# Replace a substring
df_replace = df_daft_pylist.with_column('text_replaced', F.replace(df_daft_pylist['text'], 'string', 'TEXT'))
print("\nStrings with replacement:")
df_replace.show()
```

## Testing String Functions that aren't available

```python
df_bit_length = df_daft_pylist.with_column('bit_length', F.bit_length(df_daft_pylist['text']))
print("\nBit lengths:")
df_bit_length.show()
```
